```python
def InitialLocations(location1, location2):
    if location1 == 0:
        location1 = math.floor(np.random.randint(200, 800))
        if location1 <= 0:
            location1 = 1
    if location2 == 0:
        location2 = math.floor(np.random.randint(801, 1400))
        if location2 <= location1:
            location2 = location1 + 10
        if location2 >= 2000:
            location2 = 1999
    return [location1, location2]
```

```python
def InitialRadius(radius):
    if radius == 0:
        radius = math.floor(np.random.randint(50, 150))
    return radius
```

```python
def InitialCoords(radius):
    x = math.floor(np.random.randint(150, 350))
    y = radius + 1
    if radius >= x:
        x = radius + 1
    if 500 - radius <= x:
        x = 500 - radius - 1
    return [x, y]
```
